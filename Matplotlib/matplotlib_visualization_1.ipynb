```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My First Data Visualization')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes_2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes_1.plot(x,y)
axes_2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x,y)
# plt.tight_layout()
# for axe in axes:
#     axe.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=3, ls='-', marker='o')
# ax.plot(x,x**3, label='X Cubed')
# axes[0].plot(x,y)
# axes[1].plot(y,x)
# ax.legend(loc=10)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.tight_layout()
```
